Output: -1

- Explanation: the maximum subarray is [-1], its sum is -1

```python
def max_subarray(nums):
    """
    Finds the contiguous subarray (containing at least one number) which has the largest sum
    and returns its sum along with the subarray.
    
    :param nums: List[int] - list of integers
    :return: Tuple[int, List[int]] - the sum of the maximum subarray and the subarray itself
    """
    max_sum = float('-inf')  # Initialize the maximum sum to negative infinity
    start = end = 0  # Initialize the starting and ending indices of the max subarray
    
    # Loop through each starting point of the subarray
    for i in range(len(nums)):
        current_sum = 0  # Initialize the current sum
        
        # Loop through each ending point of the subarray
        for j in range(i, len(nums)):
            current_sum += nums[j]  # Add the current element to the current sum
            
            # Update the maximum sum and indices if the current sum is greater
            if current_sum > max_sum:
                max_sum = current_sum
                start = i
                end = j
    
    return max_sum, nums[start:end+1]

# Test case
nums = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
max_sum, subarray = max_subarray(nums)
print(f"Maximum Sum: {max_sum}")  # Expected output: 6
print(f"Subarray: {subarray}")    # Expected output: [4, -1, 2, 1]
```

#### Solve the problem [Python]

- Example 1:

Explanation: the longest substring without repeating characters is "cdbegh", its length is 6

- Example 2:

Input: str = "aaaaa"

Explanation: the longest substring without repeating characters is "a", its length is 1

- Example 3:

Input: str = "eddy"

Output: 2

Explanation: the longest substring without repeating characters is "ed" (or "dy"), its length is 2

```python
def longest_substring_without_repeating(s):
    # Dictionary to store the last position of each character
    char_index = {}
    start = 0  # Start index of the current window
    max_length = 0  # Maximum length of substring found
    longest_substr = ""  # The longest substring without repeating characters

    for end in range(len(s)):
        current_char = s[end]

        # If the character is in the dictionary and its index is within the current window
        if current_char in char_index and char_index[current_char] >= start:
            start = char_index[current_char] + 1  # Move the start to the right of the last seen index

        char_index[current_char] = end  # Update the last seen index of the character

        # Calculate the length of the current window
        window_length = end - start + 1

        # Update the maximum length and longest substring if needed
        if window_length > max_length:
            max_length = window_length
            longest_substr = s[start:end + 1]

    return longest_substr, max_length

# Example usage
# s = "abcabcbb"
s = "abcdbeghef"

longest_substr, max_length = longest_substring_without_repeating(s)
print("Longest substring without repeating characters is:", longest_substr)
print("Length of this substring is:", max_length)
```

```python
def minimum_cost_path(matrix):
    n = len(matrix)
    m = len(matrix[0])
    
    minimumCostMatrix = [[0 for _ in range(m)] for _ in range(n)]
    minimumCostMatrix[0][0] = matrix[0][0]

    # Initialize the first column of the cost matrix
    for i in range(1, n):
        minimumCostMatrix[i][0] = minimumCostMatrix[i-1][0] + matrix[i][0]

    # Initialize the first row of the cost matrix
    for j in range(1, m):
        minimumCostMatrix[0][j] = minimumCostMatrix[0][j-1] + matrix[0][j]

    # Fill the rest of the cost matrix
    for i in range(1, n):
        for j in range(1, m):
            minimumCostMatrix[i][j] = matrix[i][j] + min(
                minimumCostMatrix[i-1][j],  # Cost from above
                minimumCostMatrix[i][j-1],  # Cost from left
                minimumCostMatrix[i-1][j-1] # Cost from diagonal
            )

    return minimumCostMatrix

# Example usage
matrix = [[3, 12, 4, 7, 10], [6, 8, 15, 11, 4], [19, 5, 14, 32, 21], [1, 20, 2, 9, 7]]
result = minimum_cost_path(matrix)

for row in result:
    print(row)
```

The "boat to save people" problem, often referred to as the "rescue boat problem" or "ferrying problem," involves transporting people across a river with constraints such as boat capacity and individual weights. A typical formulation is: Given a list of people's weights and a boat with a maximum weight capacity, determine the minimum number of boat trips required to transport everyone across the river.

Here's a step-by-step solution in Python using a greedy algorithm:

Sort the list of weights: This helps in efficiently pairing the heaviest and lightest person.
Use two pointers: One starting from the lightest person (beginning of the list) and one from the heaviest person (end of the list).
Pair the heaviest and lightest person: If their combined weight is within the boat's capacity, move both pointers inward. If not, move only the pointer for the heaviest person.
Count the trips: Each successful or failed pairing counts as one trip.
Here's the implementation:

```python
def num_rescue_boats(people, limit):
    people.sort()
    i, j = 0, len(people) - 1
    boats = 0
    
    while i <= j:
        if (people[i] + people[j]) <= limit:
            i += 1  # move the pointer for the lightest person
        j -= 1  # move the pointer for the heaviest person
        boats += 1
    
    return boats

# Example usage:
people = [70, 50, 80, 50]
limit = 100
print(num_rescue_boats(people, limit))  # Output: 3
```

# Candy Distribution Problem:

Problem: Given N children standing in a line, each child is given a rating value. You are required to distribute candies to these children such that each child must have at least one candy and children with a higher rating get more candies than their neighbors. Determine the minimum number of candies you need to distribute.
Example: ratings = [1, 0, 2] should output 5.


```python
def candy(ratings):
    n = len(ratings)
    candies = [1] * n
    
    for i in range(1, n):
        if ratings[i] > ratings[i-1]:
            candies[i] = candies[i-1] + 1
    
    for i in range(n-2, -1, -1):
        if ratings[i] > ratings[i+1]:
            candies[i] = max(candies[i], candies[i+1] + 1)
    
    return sum(candies)

# Example usage:
# ratings = [1, 0, 2]
ratings = [1, 3, 4, 5, 2]

print(candy(ratings))  # Output: 5
```

# Meeting Room Scheduling:

Problem: Given an array of meeting time intervals, determine if a person can attend all meetings.
Example: intervals = [[0, 30], [5, 10], [15, 20]] should output False.

```python
import heapq

def minMeetingRooms(intervals):
    if not intervals:
        return 0
    
    intervals.sort(key=lambda x: x[0])
    heap = []
    
    heapq.heappush(heap, intervals[0][1])
    
    for interval in intervals[1:]:
        if interval[0] >= heap[0]:
            heapq.heappop(heap)
        heapq.heappush(heap, interval[1])
        
    return len(heap)

# Example usage:
intervals = [[0, 30], [5, 10], [15, 20]]
print(minMeetingRooms(intervals))  # Output: 2
```

Task Scheduler:

Problem: Given a list of tasks and a cooldown period n, find the minimum intervals required to finish all tasks.
Example: tasks = ["A", "A", "A", "B", "B", "B"], n = 2 should output 8.

```python
from collections import Counter

def leastInterval(tasks, n):
    task_counts = Counter(tasks)
    max_tasks = max(task_counts.values())
    max_count = list(task_counts.values()).count(max_tasks)
    
    intervals = (max_tasks - 1) * (n + 1) + max_count
    return max(intervals, len(tasks))

# Example usage:
tasks = ["A", "A", "A", "B", "B", "B"]
n = 2
print(leastInterval(tasks, n))  # Output: 8
```

# Gas Station Problem:

Problem: There are n gas stations along a circular route, where the amount of gas at station i is gas[i]. You have a car with an unlimited gas tank and it costs cost[i] of gas to travel from station i to its next station (i+1). Determine the starting gas station index if you can travel around the circuit once in the clockwise direction, otherwise return -1.
Example: gas = [1,2,3,4,5], cost = [3,4,5,1,2] should output 3.

```python
def canCompleteCircuit(gas, cost):
    total_tank = 0
    current_tank = 0
    start_station = 0
    
    for i in range(len(gas)):
        total_tank += gas[i] - cost[i]
        current_tank += gas[i] - cost[i]
        
        if current_tank < 0:
            start_station = i + 1
            current_tank = 0
            
    return start_station if total_tank >= 0 else -1

# Example usage:
gas = [1, 2, 3, 4, 5]
cost = [3, 4, 5, 1, 2]
print(canCompleteCircuit(gas, cost))  # Output: 3
```

# Merge Intervals:

Problem: Given a collection of intervals, merge all overlapping intervals.
Example: intervals = [[1,3],[2,6],[8,10],[15,18]] should output [[1,6],[8,10],[15,18]].


```python
def merge(intervals):
    if not intervals:
        return []
    
    intervals.sort(key=lambda x: x[0])
    merged = [intervals[0]]
    
    for i in range(1, len(intervals)):
        if intervals[i][0] <= merged[-1][1]:
            merged[-1][1] = max(merged[-1][1], intervals[i][1])
        else:
            merged.append(intervals[i])
            
    return merged

# Example usage:
intervals = [[1, 3], [2, 6], [8, 10], [15, 18]]
print(merge(intervals))  # Output: [[1, 6], [8, 10], [15, 18]]
```

# Minimum Number of Platforms:

Problem: Given arrival and departure times of trains on a single day, find the minimum number of platforms required at the railway station so that no train waits.
Example: arrivals = [900, 940, 950, 1100, 1500, 1800], departures = [910, 1200, 1120, 1130, 1900, 2000] should output 3.


```python
def findPlatform(arrivals, departures):
    arrivals.sort()
    departures.sort()
    
    platform_needed = 1
    result = 1
    i = 1
    j = 0
    
    while i < len(arrivals) and j < len(departures):
        if arrivals[i] <= departures[j]:
            platform_needed += 1
            i += 1
        elif arrivals[i] > departures[j]:
            platform_needed -= 1
            j += 1
        
        if platform_needed > result:
            result = platform_needed
            
    return result

# Example usage:
arrivals = [900, 940, 950, 1100, 1500, 1800]
departures = [910, 1200, 1120, 1130, 1900, 2000]
print(findPlatform(arrivals, departures))  # Output: 3
```

# Greedy Algo- Fractional Knapsack problem 

### Problem Statement
Given weights and values of n items, the goal is to put these items in a knapsack of capacity W to get the maximum total value in the knapsack. You are allowed to break items for maximizing the total value of the knapsack.

#### Example
Suppose we have the following items:

- Item 1: Weight = 10, Value = 60
- Item 2: Weight = 20, Value = 100
- Item 3: Weight = 30, Value = 120

or

- weights=[10,20,30], values=[60,100,120]

Knapsack capacity = 50.

### Solution
To solve the Fractional Knapsack problem, follow these steps:

- Compute the value-to-weight ratio for each item.
- Sort the items based on this ratio in descending order.
- Add items to the knapsack starting from the item with the highest ratio until the knapsack is full or all items are considered.
- If adding the entire item exceeds the capacity, take the fraction of the item that fits.

# 0/1 Knapsack problem

### Problem Statement
Given weights and values of n items, put these items in a knapsack of capacity W to get the maximum total value in the knapsack. You cannot break items; either pick the whole item or don't pick it.

- Example
Suppose we have the following items:

- Item 1: Weight = 10, Value = 60
- Item 2: Weight = 20, Value = 100
- Item 3: Weight = 30, Value = 120
- Knapsack capacity = 50.

## Solution
The 0/1 Knapsack problem can be solved using dynamic programming. The idea is to build a 2D table dp where dp[i][w] represents the maximum value that can be achieved with the first i items and a knapsack capacity of w.

Here is the Python code to solve the 0/1 Knapsack problem:

```python
def knapsack_01(capacity, weights, values, n):
    # Create a 2D array to store the maximum value at each n and capacity
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]

    # Build the dp array from the bottom up
    for i in range(n + 1):
        for w in range(capacity + 1):
            if i == 0 or w == 0:
                dp[i][w] = 0
            elif weights[i - 1] <= w:
                dp[i][w] = max(values[i - 1] + dp[i - 1][w - weights[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]

# Example usage
values = [60, 100, 120]
weights = [10, 20, 30]
capacity = 50
n = len(values)
max_value = knapsack_01(capacity, weights, values, n)
print(f"Maximum value in the knapsack: {max_value}")

######-------------------------------------------------------------------------------
#####------------------ Recusive approch::::::-------------------------------
def knapsack_recursive(weights, values, W, n, memo):
    # Base Case: No items left or no capacity left
    if n == 0 or W == 0:
        return 0

    # Check if the result is already in the memo table
    if memo[n][W] != -1:
        return memo[n][W]

    # If the weight of the nth item is more than the capacity W, it cannot be included
    if weights[n-1] > W:
        memo[n][W] = knapsack_recursive(weights, values, W, n-1, memo)
        return memo[n][W]
    else:
        # Calculate the maximum value by including or excluding the nth item
        include_item = values[n-1] + knapsack_recursive(weights, values, W - weights[n-1], n-1, memo)
        exclude_item = knapsack_recursive(weights, values, W, n-1, memo)
        memo[n][W] = max(include_item, exclude_item)
        return memo[n][W]

def knapsack(weights, values, W):
    n = len(weights)
    memo = [[-1 for _ in range(W + 1)] for _ in range(n + 1)]
    return knapsack_recursive(weights, values, W, n, memo)

# Example usage
values = [60, 100, 120]
weights = [10, 20, 30]
capacity = 50
max_value = knapsack(weights, values, capacity)
print(f"Maximum value in the knapsack: {max_value}")
```

# Coin Change Problem:

Problem: Given a list of coin denominations and a total amount of money, determine the minimum number of coins required to make up that amount.
Example: coins = [1, 2, 5], amount = 11 should output 3.

```python
def coinChange(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    
    for coin in coins:
        for x in range(coin, amount + 1):
            dp[x] = min(dp[x], dp[x - coin] + 1)
    
    return dp[amount] if dp[amount] != float('inf') else -1

# Example usage:
coins = [1, 2, 5]
amount = 11
print(coinChange(coins, amount))  # Output: 3
```

# Longest Consecutive Sequence:

Problem: Given an unsorted array of integers, find the length of the longest consecutive elements sequence.
Example: nums = [100, 4, 200, 1, 3, 2] should output 4.


```python
def longestConsecutive(nums):
    num_set = set(nums)
    longest_streak = 0
    
    for num in nums:
        if num - 1 not in num_set:
            current_num = num
            current_streak = 1
            
            while current_num + 1 in num_set:
                current_num += 1
                current_streak += 1
                
            longest_streak = max(longest_streak, current_streak)
    
    return longest_streak

# Example usage:
nums = [100, 4, 200, 1, 3, 2]
print(longestConsecutive(nums))  # Output: 4
```

# Unique Paths in a Grid:

Problem: Given a grid of size m x n, find the number of unique paths from the top-left corner to the bottom-right corner, assuming you can only move right or down.
Example: m = 3, n = 2 should output 3.


```python
def uniquePaths(m, n):
    dp = [[1] * n for _ in range(m)]
    
    for i in range(1, m):
        for j in range(1, n):
            dp[i][j] = dp[i - 1][j] + dp[i][j - 1]
    
    return dp[m - 1][n - 1]

# Example usage:
m = 3
n = 2
print(uniquePaths(m, n))  # Output: 3
```
